```
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Display dataset structure
print("Dataset Structure:")
print(df.info())

# First few rows
print("\nFirst 5 Rows:")
print(df.head())
```

```
import matplotlib.pyplot as plt

# Plot histograms for numerical features
df.hist(figsize=(10, 8), bins=20, color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Numerical Features', fontsize=16)
plt.show()
```

```
# Bar plot for categorical data (species distribution)
species_counts = df['species'].value_counts()
species_counts.plot(kind='bar', color=['blue', 'green', 'red'])
plt.title('Species Distribution')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```
# Pie chart for species proportions
species_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'pink'], figsize=(6, 6))
plt.title('Proportion of Species')
plt.ylabel('')
plt.show()
```

```
import seaborn as sns

# Box plot for numerical features grouped by species
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='species', y='sepal length (cm)')
plt.title('Box Plot of Sepal Length by Species')
plt.show()
```

```
# Violin plot for numerical features grouped by species
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='species', y='sepal width (cm)', palette='muted')
plt.title('Violin Plot of Sepal Width by Species')
plt.show()
```

```
# Pair plot for all numerical features
sns.pairplot(df, hue='species', diag_kind='hist', palette='bright')
plt.suptitle('Pair Plot of Numerical Features', y=1.02)
plt.show()
```

```
# Heatmap of the correlation matrix
correlation_matrix = df.iloc[:, :-1].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of Feature Correlations')
plt.show()
```

```
sns.set_style('whitegrid')
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='species', y='petal length (cm)', palette='pastel')
plt.title('Customized Box Plot of Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```
